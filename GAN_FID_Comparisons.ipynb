```python
import matplotlib.pyplot as plt
```

```python
fid_file = "fid_scores.txt"

with open('./output_DCGAN/'+fid_file, 'r') as file:
    DCGAN = [float(line.strip()) for line in file.readlines()]

with open('./output_WGAN/'+fid_file, 'r') as file:
    WGAN = [float(line.strip()) for line in file.readlines()]

with open('./output_ACGAN/'+fid_file, 'r') as file:
    ACGAN = [float(line.strip()) for line in file.readlines()]
```

```python
plt.figure(figsize=(10,5))
plt.title("Comparing FID Scores DCGAN vs WGAN vs ACGAN")
plt.plot(DCGAN,label="DCGAN", color='blue')
plt.plot(WGAN,label="WGAN", color='orange')
plt.plot(ACGAN,label="ACGAN", color="red")
plt.xlabel("Epochs")
plt.ylabel("FID Score")
plt.legend()
plt.savefig('GAN_FID_Comparisons.jpg',
            format='jpeg')
plt.show()
```
